````python
# Step 2.2: Base Model Training and Evaluation
#
# This script focuses on training several baseline regression models using the preprocessed data.
# We will evaluate their performance using standard metrics and visualize their predictions
# and feature importances (where applicable).

# ## 1. Setup and Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import logging
from datetime import datetime
import time

# Scikit-learn imports
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - [%(filename)s:%(lineno)d] - %(message)s', force=True)

# Plotting style
try:
    plt.style.use('seaborn-v0_8-whitegrid')
except OSError:
    plt.style.use('seaborn-whitegrid')
sns.set_palette('viridis')

logging.info("Libraries imported and logging configured.")

# ## 2. Configuration and Data Loading
#
# Load the preprocessed training and testing datasets generated by step_2_1_preprocessing_pipeline.py.

logging.info("Setting up configuration for data loading.")
try:
    SCRIPT_DIR = os.path.dirname(os.path.abspath(__file__))
    WORKSPACE_ROOT = os.path.abspath(os.path.join(SCRIPT_DIR, '../../../'))
    if os.path.basename(WORKSPACE_ROOT) != 'Alpha GO':
        logging.warning(f"Auto-detected WORKSPACE_ROOT might be incorrect: {WORKSPACE_ROOT}. Falling back.")
        raise ValueError("Incorrect workspace root detected")
except Exception:
    logging.warning(f"Could not auto-detect WORKSPACE_ROOT correctly. Using hardcoded path as fallback.")
    WORKSPACE_ROOT = "D:/SBT-JAPAN/Alpha GO" #! IMPORTANT: UPDATE THIS TO YOUR ABSOLUTE WORKSPACE PATH IF NEEDED

PREPROCESSED_DATA_DIR = os.path.join(WORKSPACE_ROOT, "dataset-consolidation", "data", "final")
logging.info(f"WORKSPACE_ROOT set to: {WORKSPACE_ROOT}")
logging.info(f"PREPROCESSED_DATA_DIR set to: {PREPROCESSED_DATA_DIR}")

def find_latest_file_by_keyword(directory, keyword):
    """Finds the most recent file containing a specific keyword in its name within the directory."""
    logging.debug(f"Searching for files with keyword '{keyword}' in directory: {directory}")
    try:
        if not os.path.isdir(directory):
            logging.error(f"Directory not found: {directory}")
            return None
        # Find files containing the keyword and ending with .csv
        files = [f for f in os.listdir(directory) if keyword in f and f.endswith(".csv") and os.path.isfile(os.path.join(directory, f))]
        if not files:
            logging.error(f"No files found with keyword '{keyword}' in {directory}")
            return None
        # Sort by full name (assuming YYYYMMDD_HHMMSS prefix for chronological order)
        files.sort(reverse=True)
        latest_file = files[0]
        logging.debug(f"Found latest file for keyword '{keyword}': {latest_file}")
        return os.path.join(directory, latest_file)
    except Exception as e:
        logging.error(f"Error finding latest file with keyword '{keyword}': {e}", exc_info=True)
        return None

logging.info("Looking for latest processed data files...")
# Use keywords that are unique parts of the filenames saved by the previous script
X_train_path = find_latest_file_by_keyword(PREPROCESSED_DATA_DIR, 'X_train_processed')
X_test_path = find_latest_file_by_keyword(PREPROCESSED_DATA_DIR, 'X_test_processed')
y_train_path = find_latest_file_by_keyword(PREPROCESSED_DATA_DIR, 'y_train.csv') # More specific to avoid X_train
y_test_path = find_latest_file_by_keyword(PREPROCESSED_DATA_DIR, 'y_test.csv')   # More specific to avoid X_test

# Ensure y_train_path and y_test_path are not accidentally picking up X_train/X_test if keywords are too broad
# A more robust way for y_train and y_test is to ensure they don't contain "X_"
def find_latest_y_file(directory, base_keyword): # e.g. base_keyword = "y_train"
    logging.debug(f"Searching for Y files with base keyword '{base_keyword}' in directory: {directory}")
    try:
        if not os.path.isdir(directory):
            logging.error(f"Directory not found: {directory}")
            return None
        # Files should contain base_keyword, end with .csv, and importantly NOT contain "X_"
        files = [f for f in os.listdir(directory) if base_keyword in f and f.endswith(".csv") and "X_" not in f and os.path.isfile(os.path.join(directory,f))]
        if not files:
            logging.error(f"No Y files found with base keyword '{base_keyword}' in {directory}")
            return None
        files.sort(reverse=True)
        latest_file = files[0]
        logging.debug(f"Found latest Y file for base keyword '{base_keyword}': {latest_file}")
        return os.path.join(directory, latest_file)
    except Exception as e:
        logging.error(f"Error finding latest Y file for '{base_keyword}': {e}", exc_info=True)
        return None

# Re-fetch y_train and y_test with the more specific finder
y_train_path = find_latest_y_file(PREPROCESSED_DATA_DIR, 'y_train')
y_test_path = find_latest_y_file(PREPROCESSED_DATA_DIR, 'y_test')


if not all([X_train_path, X_test_path, y_train_path, y_test_path]):
    logging.critical("One or more processed data files not found. Please check the PREPROCESSED_DATA_DIR and ensure step_2_1 ran successfully.")
    print("CRITICAL: Data files not found. Script execution cannot continue as expected.")
    X_train_df, X_test_df, y_train_series, y_test_series = pd.DataFrame(), pd.DataFrame(), pd.Series(dtype='float64'), pd.Series(dtype='float64')
else:
    logging.info(f"Loading X_train from: {X_train_path}")
    X_train_df = pd.read_csv(X_train_path)
    logging.info(f"Loading X_test from: {X_test_path}")
    X_test_df = pd.read_csv(X_test_path)
    logging.info(f"Loading y_train from: {y_train_path}")
    y_train_series = pd.read_csv(y_train_path).squeeze()
    logging.info(f"Loading y_test from: {y_test_path}")
    y_test_series = pd.read_csv(y_test_path).squeeze()

    logging.info("Data loading complete.")
    print(f"X_train shape: {X_train_df.shape}")
    print(f"X_test shape: {X_test_df.shape}")
    print(f"y_train shape: {y_train_series.shape}")
    print(f"y_test shape: {y_test_series.shape}")

    if X_train_df.shape[0] != y_train_series.shape[0]:
        logging.error("Mismatch in number of samples between X_train and y_train!")
    if X_test_df.shape[0] != y_test_series.shape[0]:
        logging.error("Mismatch in number of samples between X_test and y_test!")
    if not X_train_df.empty and not X_test_df.empty and X_train_df.shape[1] != X_test_df.shape[1]:
        logging.error("Mismatch in number of features between X_train and X_test!")

    if not X_train_df.empty:
        print("\nFirst 5 rows of X_train_df:")
        print(X_train_df.head())
    if not y_train_series.empty:
        print("\nFirst 5 rows of y_train_series:")
        print(y_train_series.head())

# ... (rest of the script: Model Definitions, Training Loop, Results Summary remains the same) ...
# Make sure to copy the rest of the script from the previous response from here downwards

# ## 3. Model Definitions
logging.info("Defining baseline models.")
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(random_state=42, n_jobs=-1, n_estimators=100),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42, n_estimators=100),
    "K-Nearest Neighbors": KNeighborsRegressor(n_jobs=-1),
    "Support Vector Regressor": SVR(C=1.0, epsilon=0.1)
}
model_results = []
logging.info(f"Models defined: {list(models.keys())}")

# ## 4. Model Training and Evaluation Loop
def train_and_evaluate_model(model, X_train, y_train, X_test, y_test, model_name):
    """Trains a model, makes predictions, evaluates, and visualizes."""
    logging.info(f"Initiating training for {model_name}...")
    start_time = time.time()
    try:
        model.fit(X_train, y_train)
    except Exception as e:
        logging.error(f"Error during training {model_name}: {e}", exc_info=True)
        return {
            "Model": model_name, "R-squared": np.nan, "MAE": np.nan,
            "RMSE": np.nan, "Training Time (s)": np.nan, "Status": "Training Failed"
        }
    end_time = time.time()
    training_time = end_time - start_time
    logging.info(f"Training for {model_name} completed in {training_time:.2f} seconds.")

    logging.info(f"Making predictions with {model_name} on test data...")
    try:
        predictions = model.predict(X_test)
    except Exception as e:
        logging.error(f"Error during prediction with {model_name}: {e}", exc_info=True)
        return {
            "Model": model_name, "R-squared": np.nan, "MAE": np.nan,
            "RMSE": np.nan, "Training Time (s)": training_time, "Status": "Prediction Failed"
        }

    logging.info(f"Calculating metrics for {model_name}...")
    try:
        r2 = r2_score(y_test, predictions)
        mae = mean_absolute_error(y_test, predictions)
        rmse = np.sqrt(mean_squared_error(y_test, predictions))
    except Exception as e:
        logging.error(f"Error calculating metrics for {model_name}: {e}", exc_info=True)
        r2, mae, rmse = np.nan, np.nan, np.nan

    print(f"\n--- {model_name} Performance ---")
    print(f"R-squared: {r2:.4f}")
    print(f"MAE: {mae:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"Training Time: {training_time:.2f}s")

    results_dict = {
        "Model": model_name, "R-squared": r2, "MAE": mae, "RMSE": rmse,
        "Training Time (s)": training_time, "Status": "Completed"
    }

    logging.info(f"Generating Actual vs. Predicted plot for {model_name}.")
    plt.figure(figsize=(8, 6))
    plt.scatter(y_test, predictions, alpha=0.5, edgecolors='k', s=50, label='Predictions')
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2, label='Ideal Fit')
    plt.xlabel("Actual Price Log")
    plt.ylabel("Predicted Price Log")
    plt.title(f"{model_name}: Actual vs. Predicted Values (R2: {r2:.3f})")
    plt.legend()
    plt.show()

    if hasattr(model, 'feature_importances_'):
        logging.info(f"Generating Feature Importances plot for {model_name}.")
        importances = model.feature_importances_
        if X_train.shape[1] == len(importances):
            feature_names = X_train.columns
            feature_importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances})
            feature_importance_df = feature_importance_df.sort_values(by='importance', ascending=False)

            top_n = 20
            plt.figure(figsize=(10, max(6, top_n * 0.35)))
            sns.barplot(x='importance', y='feature', data=feature_importance_df.head(top_n), palette='viridis_r')
            plt.title(f"Top {top_n} Feature Importances for {model_name}")
            plt.tight_layout()
            plt.show()
            results_dict['Top_Features'] = feature_importance_df.head(top_n)['feature'].tolist()
        else:
            logging.warning(f"Mismatch between X_train features ({X_train.shape[1]}) and importances ({len(importances)}) for {model_name}.")
            
    return results_dict

logging.info("Starting model training and evaluation loop.")
if 'X_train_df' in locals() and not X_train_df.empty and \
   'y_train_series' in locals() and not y_train_series.empty and \
   'X_test_df' in locals() and not X_test_df.empty and \
   'y_test_series' in locals() and not y_test_series.empty:
    for model_name, model_instance in models.items():
        logging.info(f"Processing model: {model_name}")
        result = train_and_evaluate_model(model_instance, X_train_df, y_train_series, X_test_df, y_test_series, model_name)
        model_results.append(result)
        logging.info(f"Finished processing for model: {model_name}. Result: {result.get('Status', 'Unknown')}")
else:
    logging.critical("Training data not loaded or empty. Skipping model training loop.")

logging.info("Model training and evaluation loop finished.")

# ## 5. Results Summary
logging.info("Summarizing model performance results.")
if model_results:
    results_df = pd.DataFrame(model_results)
    results_df['R-squared'] = pd.to_numeric(results_df['R-squared'], errors='coerce')
    results_df = results_df.sort_values(by='R-squared', ascending=False, na_position='last').reset_index(drop=True)

    print("\n--- Model Performance Summary ---")
    print(results_df)

    completed_models_df = results_df[results_df['Status'] == 'Completed']
    if not completed_models_df.empty and 'Model' in completed_models_df.columns and 'R-squared' in completed_models_df.columns:
        best_model_name = completed_models_df.iloc[0]['Model']
        best_r2 = completed_models_df.iloc[0]['R-squared']
        logging.info(f"Best performing model (Completed) based on R-squared: {best_model_name} (R2: {best_r2:.4f})")
        print(f"\nBest performing model (Completed) based on R-squared: {best_model_name} (R2: {best_r2:.4f})")
    elif not completed_models_df.empty:
        logging.warning("Could not determine best model due to missing columns in completed_models_df.")
        print("\nCould not reliably determine best model from completed models.")
    else:
        logging.warning("No models completed successfully to determine the best one.")
        print("\nNo models completed successfully.")
else:
    logging.warning("model_results list is empty. No models were processed or all failed early.")
    print("\nNo model results to display.")

logging.info("Script finished.")
````

````python
# Step 3.1: Hyperparameter Tuning for Top Models
#
# This script focuses on tuning the hyperparameters of the top-performing baseline models
# (Random Forest, SVR, KNN, Gradient Boosting) using RandomizedSearchCV.
# It uses the data preprocessed with IterativeImputer for 'annual_insurance'.

import pandas as pd
import numpy as np
import os
import logging
import time
from datetime import datetime

# Scikit-learn imports
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.model_selection import RandomizedSearchCV, KFold
from sklearn.metrics import r2_score # To report final R2 on best model if needed, though CV score is primary here

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - [%(filename)s:%(lineno)d] - %(message)s', force=True)

logging.info("Libraries imported and logging configured.")
logging.info("Starting Step 3.1: Hyperparameter Tuning for Top Models.")

# --- 1. Configuration and Data Loading ---
logging.info("Setting up configuration for data loading.")
try:
    SCRIPT_DIR = os.path.dirname(os.path.abspath(__file__))
    WORKSPACE_ROOT = os.path.abspath(os.path.join(SCRIPT_DIR, '../../../'))
    if os.path.basename(WORKSPACE_ROOT) != 'Alpha GO': # Replace 'Alpha GO' with your actual workspace directory name if different
        logging.warning(f"Auto-detected WORKSPACE_ROOT might be incorrect: {WORKSPACE_ROOT}. Falling back.")
        raise ValueError("Incorrect workspace root detected")
except Exception:
    logging.warning(f"Could not auto-detect WORKSPACE_ROOT correctly. Using hardcoded path as fallback.")
    WORKSPACE_ROOT = "D:/SBT-JAPAN/Alpha GO" #! IMPORTANT: UPDATE THIS TO YOUR ABSOLUTE WORKSPACE PATH IF NEEDED

PREPROCESSED_DATA_DIR = os.path.join(WORKSPACE_ROOT, "dataset-consolidation", "data", "final")
MODELS_DIR = os.path.join(WORKSPACE_ROOT, "dataset-consolidation", "models", "base_models") # For saving tuned models
os.makedirs(MODELS_DIR, exist_ok=True)

logging.info(f"WORKSPACE_ROOT set to: {WORKSPACE_ROOT}")
logging.info(f"PREPROCESSED_DATA_DIR set to: {PREPROCESSED_DATA_DIR}")
logging.info(f"MODELS_DIR set to: {MODELS_DIR}")

def find_latest_file_by_keyword(directory, keyword):
    """Finds the most recent file containing a specific keyword in its name within the directory."""
    logging.debug(f"Searching for files with keyword '{keyword}' in directory: {directory}")
    try:
        if not os.path.isdir(directory):
            logging.error(f"Directory not found: {directory}")
            return None
        files = [f for f in os.listdir(directory) if keyword in f and f.endswith(".csv") and os.path.isfile(os.path.join(directory, f))]
        if not files:
            logging.error(f"No files found with keyword '{keyword}' in {directory}")
            return None
        files.sort(reverse=True) # Assumes YYYYMMDD_HHMMSS prefix for chronological order
        latest_file = files[0]
        logging.debug(f"Found latest file for keyword '{keyword}': {latest_file}")
        return os.path.join(directory, latest_file)
    except Exception as e:
        logging.error(f"Error finding latest file with keyword '{keyword}': {e}", exc_info=True)
        return None

def find_latest_y_file(directory, base_keyword):
    logging.debug(f"Searching for Y files with base keyword '{base_keyword}' in directory: {directory}")
    try:
        if not os.path.isdir(directory):
            logging.error(f"Directory not found: {directory}")
            return None
        files = [f for f in os.listdir(directory) if base_keyword in f and f.endswith(".csv") and "X_" not in f and os.path.isfile(os.path.join(directory,f))]
        if not files:
            logging.error(f"No Y files found with base keyword '{base_keyword}' in {directory}")
            return None
        files.sort(reverse=True)
        latest_file = files[0]
        logging.debug(f"Found latest Y file for base keyword '{base_keyword}': {latest_file}")
        return os.path.join(directory, latest_file)
    except Exception as e:
        logging.error(f"Error finding latest Y file for '{base_keyword}': {e}", exc_info=True)
        return None

logging.info("Looking for latest processed data files (from IterativeImputer run)...")
X_train_path = find_latest_file_by_keyword(PREPROCESSED_DATA_DIR, 'X_train_processed')
# X_test_path = find_latest_file_by_keyword(PREPROCESSED_DATA_DIR, 'X_test_processed') # Test set not strictly needed for CV tuning
y_train_path = find_latest_y_file(PREPROCESSED_DATA_DIR, 'y_train')

if not all([X_train_path, y_train_path]):
    logging.critical("Training data files (X_train_processed, y_train) not found. Please ensure step_2_1 (with IterativeImputer) ran successfully.")
    print("CRITICAL: Training data files not found. Script cannot continue.")
    # Exit if files are not found
    exit()
else:
    logging.info(f"Loading X_train from: {X_train_path}")
    X_train_df = pd.read_csv(X_train_path)
    logging.info(f"Loading y_train from: {y_train_path}")
    y_train_series = pd.read_csv(y_train_path).squeeze()

    logging.info("Training data loading complete.")
    print(f"X_train shape: {X_train_df.shape}")
    print(f"y_train shape: {y_train_series.shape}")

    if X_train_df.shape[0] != y_train_series.shape[0]:
        logging.error("Mismatch in number of samples between X_train and y_train! Exiting.")
        exit()
    if X_train_df.empty or y_train_series.empty:
        logging.error("Training data or target is empty! Exiting.")
        exit()

# --- 2. Define Models and Hyperparameter Grids ---
logging.info("Defining models and hyperparameter grids for RandomizedSearchCV.")

# Reduced n_estimators for faster RandomizedSearch; can be increased for final model
param_grids = {
    "Random Forest": {
        'n_estimators': [100, 200, 300], # Reduced for quicker search
        'max_features': ['sqrt', 'log2', 0.5, 0.7], # String options for older sklearn, float for proportion
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'bootstrap': [True, False]
    },
    "Gradient Boosting": {
        'n_estimators': [100, 200, 300], # Reduced
        'learning_rate': [0.01, 0.05, 0.1, 0.2],
        'max_depth': [3, 5, 7],
        'subsample': [0.7, 0.8, 0.9, 1.0],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    },
    "K-Nearest Neighbors": {
        'n_neighbors': list(range(3, 16)), # Test a range of neighbors
        'weights': ['uniform', 'distance'],
        'p': [1, 2] # 1 for Manhattan, 2 for Euclidean
    },
    "Support Vector Regressor": { # SVR can be slow, so smaller grid or fewer n_iter
        'C': [0.1, 1, 10, 50], # Removed 100 for speed
        'epsilon': [0.01, 0.1, 0.2],
        'kernel': ['rbf', 'linear'], # Poly can be very slow
        'gamma': ['scale', 'auto'] # Relevant for 'rbf' kernel
    }
}

models_to_tune = {
    "Random Forest": RandomForestRegressor(random_state=42, n_jobs=-1),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),
    "K-Nearest Neighbors": KNeighborsRegressor(n_jobs=-1),
    "Support Vector Regressor": SVR()
}

# --- 3. Hyperparameter Tuning Loop ---
N_ITER_SEARCH = 20  # Number of parameter settings that are sampled.
CV_FOLDS = 5        # Number of cross-validation folds.
SCORING_METRIC = 'r2' # R-squared is a good general metric for regression.

tuned_model_results = []
best_estimators = {}

logging.info(f"Starting hyperparameter tuning with RandomizedSearchCV. n_iter={N_ITER_SEARCH}, cv={CV_FOLDS}, scoring='{SCORING_METRIC}'.")

for model_name, model_instance in models_to_tune.items():
    logging.info(f"--- Tuning: {model_name} ---")
    if model_name not in param_grids:
        logging.warning(f"No parameter grid defined for {model_name}. Skipping.")
        continue

    param_dist = param_grids[model_name]
    
    # For SVR with linear kernel, 'gamma' is not applicable.
    # RandomizedSearchCV might warn or error. A more robust way is to have separate searches or conditional params.
    # For simplicity here, we rely on SVR ignoring irrelevant params or use a try-except.

    start_time = time.time()
    try:
        random_search = RandomizedSearchCV(
            estimator=model_instance,
            param_distributions=param_dist,
            n_iter=N_ITER_SEARCH if model_name != "Support Vector Regressor" else 10, # Fewer iterations for SVR due to speed
            cv=KFold(n_splits=CV_FOLDS, shuffle=True, random_state=42),
            scoring=SCORING_METRIC,
            random_state=42,
            n_jobs=-1, # Use all available cores
            verbose=1 # Set to 2 for more detailed output
        )
        random_search.fit(X_train_df, y_train_series)
        
        tuning_time = time.time() - start_time
        
        best_score = random_search.best_score_
        best_params = random_search.best_params_
        best_estimator = random_search.best_estimator_
        best_estimators[model_name] = best_estimator # Store the best fitted estimator

        logging.info(f"Best {SCORING_METRIC} for {model_name}: {best_score:.4f}")
        logging.info(f"Best parameters for {model_name}: {best_params}")
        logging.info(f"Tuning time for {model_name}: {tuning_time:.2f} seconds.")
        
        tuned_model_results.append({
            "Model": model_name,
            f"Best CV {SCORING_METRIC}": best_score,
            "Best Parameters": best_params,
            "Tuning Time (s)": tuning_time
        })

        # Save the best estimator
        try:
            import joblib
            model_filename = f"{datetime.now().strftime('%Y%m%d_%H%M%S')}_{model_name.replace(' ', '_')}_tuned.joblib"
            model_path = os.path.join(MODELS_DIR, model_filename)
            joblib.dump(best_estimator, model_path)
            logging.info(f"Saved tuned {model_name} to {model_path}")
        except Exception as e_save:
            logging.error(f"Error saving tuned model {model_name}: {e_save}")

    except Exception as e_search:
        tuning_time = time.time() - start_time
        logging.error(f"Error during RandomizedSearchCV for {model_name}: {e_search}", exc_info=True)
        tuned_model_results.append({
            "Model": model_name,
            f"Best CV {SCORING_METRIC}": np.nan,
            "Best Parameters": "Error during search",
            "Tuning Time (s)": tuning_time
        })

logging.info("Hyperparameter tuning loop finished.")

# --- 4. Results Summary ---
logging.info("Summarizing hyperparameter tuning results.")
if tuned_model_results:
    results_df = pd.DataFrame(tuned_model_results)
    results_df = results_df.sort_values(by=f"Best CV {SCORING_METRIC}", ascending=False, na_position='last').reset_index(drop=True)

    print("\n--- Hyperparameter Tuning Summary ---")
    print(results_df)

    if not results_df.empty and not results_df[f"Best CV {SCORING_METRIC}"].isnull().all():
        best_overall_model_details = results_df.iloc[0]
        best_model_name = best_overall_model_details['Model']
        best_overall_score = best_overall_model_details[f"Best CV {SCORING_METRIC}"]
        logging.info(f"Overall best model after tuning based on CV {SCORING_METRIC}: {best_model_name} (Score: {best_overall_score:.4f})")
        print(f"\nOverall best model after tuning based on CV {SCORING_METRIC}: {best_model_name} (Score: {best_overall_score:.4f})")
    else:
        logging.warning("No models tuned successfully or no valid scores found.")
        print("\nNo models tuned successfully or no valid scores available.")
else:
    logging.warning("Tuned model results list is empty. No models were processed or all failed.")
    print("\nNo tuning results to display.")

logging.info("Script finished.")
````

````python
# Step 3.2: Feature Selection / Dimensionality Reduction
#
# This script focuses on:
# 1. Analyzing feature importances from the best tuned Random Forest.
# 2. Applying feature selection using SelectFromModel.
# 3. Retraining and evaluating top models (Random Forest, Gradient Boosting) on the reduced feature set.
# MODIFIED: Uses get_feature_names_out() for consistent naming and remainder='drop'.

import pandas as pd
import numpy as np
import os
import logging
import time
from datetime import datetime
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

# Scikit-learn imports
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import cross_val_score, KFold, train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# --- 1. Configuration and Data Loading ---
# Assuming SCRIPT_DIR, WORKSPACE_ROOT, PREPROCESSED_DATA_DIR, MODELS_DIR,
# FEATURE_ENGINEERED_DATA_DIR are defined as in Step 3.1 or earlier in the notebook.
# Also assuming find_latest_file_by_keyword, find_latest_y_file, and find_latest_feature_engineered_file are defined.

logging.info("Starting Step 3.2: Feature Selection / Dimensionality Reduction (Updated).")

# Step 3.2: Feature Selection / Dimensionality Reduction
# ... (other imports remain the same) ...

# --- 1. Configuration and Data Loading ---
logging.info("Starting Step 3.2: Feature Selection / Dimensionality Reduction (Updated).")

# Define paths directly for Jupyter Notebook environment
WORKSPACE_ROOT = "D:/SBT-JAPAN/Alpha GO" # Your absolute workspace path
FEATURE_ENGINEERED_DATA_DIR = os.path.join(WORKSPACE_ROOT, "dataset-consolidation", "data", "processed")
PREPROCESSED_DATA_DIR = os.path.join(WORKSPACE_ROOT, "dataset-consolidation", "data", "final")
MODELS_DIR = os.path.join(WORKSPACE_ROOT, "dataset-consolidation", "models", "base_models") # For saving tuned models and selected features
# SCRIPT_DIR is not strictly needed if WORKSPACE_ROOT is well-defined, but if you need it for other purposes:
# NOTEBOOK_DIR = os.path.dirname("D:/SBT-JAPAN/Alpha GO/dataset-consolidation/notebooks/modeling/02_base_model_training_and_evaluation.ipynb") # Path to your current notebook

os.makedirs(MODELS_DIR, exist_ok=True)
os.makedirs(PREPROCESSED_DATA_DIR, exist_ok=True)
os.makedirs(FEATURE_ENGINEERED_DATA_DIR, exist_ok=True) # Ensure this exists too

logging.info(f"WORKSPACE_ROOT set to: {WORKSPACE_ROOT}")
logging.info(f"FEATURE_ENGINEERED_DATA_DIR set to: {FEATURE_ENGINEERED_DATA_DIR}")
logging.info(f"PREPROCESSED_DATA_DIR set to: {PREPROCESSED_DATA_DIR}")
logging.info(f"MODELS_DIR set to: {MODELS_DIR}")

def find_latest_file_by_keyword(directory, keyword, endswith=".csv"):
    """Finds the most recent file containing a specific keyword and extension in its name within the directory."""
    logging.debug(f"Searching for files with keyword '{keyword}' and ending with '{endswith}' in directory: {directory}")
    try:
        if not os.path.isdir(directory):
            logging.error(f"Directory not found: {directory}")
            return None
        files = [f for f in os.listdir(directory) if keyword in f and f.endswith(endswith) and os.path.isfile(os.path.join(directory, f))]
        if not files:
            logging.warning(f"No files found with keyword '{keyword}' and ending with '{endswith}' in {directory}")
            return None
        # Sort by modification time for robustness, assuming no timestamp in name or it's not consistently first
        files.sort(key=lambda f: os.path.getmtime(os.path.join(directory, f)), reverse=True)
        latest_file = files[0]
        logging.info(f"Found latest file for keyword '{keyword}': {latest_file} in {directory}")
        return os.path.join(directory, latest_file)
    except Exception as e:
        logging.error(f"Error finding latest file with keyword '{keyword}': {e}", exc_info=True)
        return None

def find_latest_y_file(directory, base_keyword):
    """Finds the most recent 'y' file (target variable) based on a keyword."""
    logging.debug(f"Searching for Y files with base keyword '{base_keyword}' in directory: {directory}")
    try:
        if not os.path.isdir(directory):
            logging.error(f"Directory not found: {directory}")
            return None
        # Example: y_train_processed_TIMESTAMP.csv or y_train_TIMESTAMP.csv
        # Adjust if your 'y' files have a different naming pattern (e.g. no "X_" prefix)
        files = [f for f in os.listdir(directory) if base_keyword in f and f.endswith(".csv") and "X_" not in f and os.path.isfile(os.path.join(directory,f))]
        if not files:
            logging.warning(f"No Y files found with base keyword '{base_keyword}' in {directory}")
            return None
        files.sort(key=lambda f: os.path.getmtime(os.path.join(directory, f)), reverse=True)
        latest_file = files[0]
        logging.info(f"Found latest Y file for base keyword '{base_keyword}': {latest_file} in {directory}")
        return os.path.join(directory, latest_file)
    except Exception as e:
        logging.error(f"Error finding latest Y file for '{base_keyword}': {e}", exc_info=True)
        return None

def find_latest_feature_engineered_file(directory, keyword_in_name="_feature_engineered"):
    """Finds the most recent feature engineered file."""
    logging.debug(f"Searching for latest feature engineered files with '{keyword_in_name}' in directory: {directory}")
    try:
        if not os.path.isdir(directory):
            logging.error(f"Directory not found: {directory}")
            return None
        files = [f for f in os.listdir(directory) if keyword_in_name in f and f.endswith(".csv") and os.path.isfile(os.path.join(directory, f))]
        if not files:
            logging.warning(f"No feature engineered files with '{keyword_in_name}' found in {directory}")
            return None
        files.sort(key=lambda f: os.path.getmtime(os.path.join(directory, f)), reverse=True)
        latest_file = files[0]
        logging.info(f"Found latest feature engineered file: {latest_file} in {directory}")
        return os.path.join(directory, latest_file)
    except Exception as e:
        logging.error(f"Error finding latest feature engineered file: {e}", exc_info=True)
        return None


# Load original full feature-engineered data
# Using the specific filename you provided from 'data/processed'
# latest_fe_file_path = os.path.join(FEATURE_ENGINEERED_DATA_DIR, "20250520_225504_cars_feature_engineered.csv")
# Or, if you want to use the helper function:
latest_fe_file_path = find_latest_feature_engineered_file(FEATURE_ENGINEERED_DATA_DIR, keyword_in_name="_cars_feature_engineered") # Adjusted keyword
if not latest_fe_file_path:
    logging.critical(f"No feature engineered CSV file found in {FEATURE_ENGINEERED_DATA_DIR} with keyword '_cars_feature_engineered'. Please check the directory and keyword. Exiting.")
    # Attempt with a more generic keyword if the specific one fails, or list files
    logging.info(f"Attempting to list files in {FEATURE_ENGINEERED_DATA_DIR}: {os.listdir(FEATURE_ENGINEERED_DATA_DIR)}")
    exit()
logging.info(f"Loading original feature-engineered data (for preprocessor fitting and consistent splitting) from: {latest_fe_file_path}")
df_original_fe = pd.read_csv(latest_fe_file_path)

# ... (the rest of the script from "Define the target column name..." onwards should follow) ...

# Define the target column name as used in df_original_fe
TARGET_COLUMN_ORIGINAL = 'price_log' # Ensure this matches the column name in your feature-engineered file
if TARGET_COLUMN_ORIGINAL not in df_original_fe.columns:
    logging.error(f"Target column '{TARGET_COLUMN_ORIGINAL}' not found in {df_original_fe.columns.tolist()} from {latest_fe_file_path}. Exiting.")
    exit()

X_original_fe = df_original_fe.drop(columns=[TARGET_COLUMN_ORIGINAL])
y_original_fe = df_original_fe[TARGET_COLUMN_ORIGINAL]
logging.info(f"Original full feature set X_original_fe shape: {X_original_fe.shape}")
logging.info(f"Original full target y_original_fe shape: {y_original_fe.shape}")

# Split X_original_fe and y_original_fe to get the X_train part
# that was used to fit the preprocessor in step_2_1_preprocessing_pipeline.py
X_train_unprocessed, _, y_train_for_split_consistency, _ = train_test_split(
    X_original_fe, y_original_fe, test_size=0.2, random_state=42
)
logging.info(f"X_train_unprocessed (for preprocessor fitting) shape: {X_train_unprocessed.shape}")

# Load y_train (target for models)
y_train_path = find_latest_y_file(PREPROCESSED_DATA_DIR, 'y_train') # Ensure this function is available
if not y_train_path:
    logging.critical(f"y_train file not found in PREPROCESSED_DATA_DIR. Exiting.")
    exit()
logging.info(f"Loading y_train (for model training) from: {y_train_path}")
y_train_series = pd.read_csv(y_train_path).squeeze()

if len(y_train_series) != len(y_train_for_split_consistency):
    logging.warning(f"Mismatch in length between loaded y_train_series ({len(y_train_series)}) and y_train from consistent split ({len(y_train_for_split_consistency)}). This might indicate an issue.")
    # Consider exiting if this is critical: exit()


# --- 2. Re-define and Fit Preprocessor to get Feature Names ---
logging.info("Defining preprocessor_for_names to match step_2_1 structure.")
numeric_features = X_train_unprocessed.select_dtypes(include=np.number).columns.tolist()
categorical_features = X_train_unprocessed.select_dtypes(include='object').columns.tolist()
logging.info(f"Numeric features for preprocessor: {numeric_features}")
logging.info(f"Categorical features for preprocessor: {categorical_features}")

preprocessor_for_names = None
# This structure MUST mirror the preprocessor used in step_2_1_preprocessing_pipeline.py
if 'annual_insurance' in numeric_features:
    numeric_transformer_main = Pipeline(steps=[('scaler', StandardScaler())])
    numeric_transformer_insurance = Pipeline(steps=[
        ('imputer', IterativeImputer(random_state=42)), # Assuming IterativeImputer is appropriate here as in the original snippet
        ('scaler', StandardScaler())
    ])
    other_numeric_features = [col for col in numeric_features if col != 'annual_insurance']
    
    # Ensure 'other_numeric_features' are actually present in X_train_unprocessed
    other_numeric_features = [col for col in other_numeric_features if col in X_train_unprocessed.columns]
    # Ensure 'annual_insurance' is actually present for its transformer
    annual_insurance_col_list = ['annual_insurance'] if 'annual_insurance' in X_train_unprocessed.columns else []


    categorical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='constant', fill_value='missing_value_runtime')),
        ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False)) # sparse_output=False for get_feature_names_out compatibility if not default
    ])
    
    transformers_list = []
    if annual_insurance_col_list:
        transformers_list.append(('num_insurance', numeric_transformer_insurance, annual_insurance_col_list))
    if other_numeric_features: # Only add if there are other numeric features
        transformers_list.append(('num_main', numeric_transformer_main, other_numeric_features))
    if categorical_features: # Only add if there are categorical features
        transformers_list.append(('cat', categorical_transformer, categorical_features))

    if not transformers_list:
        logging.error("No transformers were defined for ColumnTransformer. Check feature lists. Exiting.")
        exit()

    preprocessor_for_names = ColumnTransformer(
        transformers=transformers_list,
        remainder='drop'  # MODIFIED: Changed from 'passthrough'
    )
elif numeric_features or categorical_features: # Handles case where 'annual_insurance' is not special or not present
    logging.info("'annual_insurance' not treated as a special numeric feature for preprocessor_for_names.")
    numeric_transformer = Pipeline(steps=[('scaler', StandardScaler())])
    categorical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='constant', fill_value='missing_value_runtime')),
        ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
    ])
    
    transformers_list = []
    if numeric_features:
        transformers_list.append(('num', numeric_transformer, numeric_features))
    if categorical_features:
        transformers_list.append(('cat', categorical_transformer, categorical_features))

    if not transformers_list:
        logging.error("No transformers were defined for ColumnTransformer in the fallback case. Check feature lists. Exiting.")
        exit()
        
    preprocessor_for_names = ColumnTransformer(
        transformers=transformers_list,
        remainder='drop' # MODIFIED: Changed from 'passthrough'
    )
else:
     logging.error("Preprocessor setup for feature names failed: No numeric or categorical features found. Exiting.")
     exit()

logging.info("Fitting preprocessor_for_names on X_train_unprocessed to get transformed feature names...")
try:
    preprocessor_for_names.fit(X_train_unprocessed)
    transformed_feature_names = list(preprocessor_for_names.get_feature_names_out()) # MODIFIED: Using get_feature_names_out()
    logging.info(f"Successfully generated {len(transformed_feature_names)} transformed feature names using get_feature_names_out().")
    logging.info(f"First 10 transformed names: {transformed_feature_names[:10]}")
    logging.info(f"Last 10 transformed names: {transformed_feature_names[-10:]}")
except Exception as e:
    logging.error(f"Could not get transformed feature names from preprocessor: {e}. Ensure preprocessor is well-defined and X_train_unprocessed is not empty. Exiting.", exc_info=True)
    exit()

# Load X_train_processed (the actual data used for tuning in Step 3.1)
X_train_processed_path = find_latest_file_by_keyword(PREPROCESSED_DATA_DIR, 'X_train_processed') # Ensure this function is available
if not X_train_processed_path:
    logging.critical(f"X_train_processed.csv file not found in {PREPROCESSED_DATA_DIR}. Ensure Step 2.1 ran and produced this file. Exiting.")
    exit()
logging.info(f"Loading X_train_processed from: {X_train_processed_path}")
X_train_processed_df = pd.read_csv(X_train_processed_path)

if X_train_processed_df.shape[1] != len(transformed_feature_names):
    logging.error(f"CRITICAL MISMATCH: Column count of loaded X_train_processed ({X_train_processed_df.shape[1]}) "
                  f"does not match generated transformed feature names count ({len(transformed_feature_names)}). "
                  f"This means 'preprocessor_for_names' in Step 3.2 is NOT ALIGNED with the preprocessor from Step 2.1 "
                  f"that created 'X_train_processed.csv'. Review both preprocessor definitions. Exiting.")
    logging.error(f"Generated names: {transformed_feature_names}")
    logging.error(f"X_train_processed_df columns (first 10): {X_train_processed_df.columns[:10].tolist()}")
    exit()

# Assign the consistent, prefixed column names to the loaded DataFrame
X_train_processed_df.columns = transformed_feature_names
logging.info(f"X_train_processed_df shape after loading and naming: {X_train_processed_df.shape}")
logging.info(f"Column names assigned to X_train_processed_df (first 10): {X_train_processed_df.columns[:10].tolist()}")


# --- 3. Load Tuned Random Forest Model ---
# Find the model tuned in Step 3.1
tuned_rf_model_path = find_latest_file_by_keyword(MODELS_DIR, "Random_Forest_tuned", endswith=".joblib") # Ensure this function is available
if not tuned_rf_model_path:
    logging.critical("Tuned Random Forest model (from Step 3.1) not found. Please run Step 3.1 first. Exiting.")
    exit()
logging.info(f"Loading tuned Random Forest model from: {tuned_rf_model_path}")
try:
    tuned_rf_model = joblib.load(tuned_rf_model_path)
except Exception as e:
    logging.error(f"Error loading tuned Random Forest model: {e}. Exiting.", exc_info=True)
    exit()

# --- 4. Feature Importances from Tuned Random Forest ---
logging.info("Extracting feature importances from tuned Random Forest.")
if not hasattr(tuned_rf_model, 'feature_importances_'):
    logging.error("Loaded model does not have 'feature_importances_'. Ensure it's a tree-based model and was fitted. Exiting.")
    exit()
importances = tuned_rf_model.feature_importances_

if len(importances) != len(X_train_processed_df.columns):
    logging.error(f"Mismatch between number of importances ({len(importances)}) and number of columns in X_train_processed_df ({len(X_train_processed_df.columns)}). Exiting.")
    exit()

feature_importance_df = pd.DataFrame({
    'feature': X_train_processed_df.columns, # Using the new prefixed names
    'importance': importances
}).sort_values(by='importance', ascending=False)

logging.info(f"Top 20 feature importances:\n{feature_importance_df.head(20)}")

plt.figure(figsize=(12, max(10, len(feature_importance_df.head(30)) // 2) )) # Adjust height for more features
sns.barplot(x='importance', y='feature', data=feature_importance_df.head(30), palette='viridis_r') # Show top 30
plt.title('Top 30 Feature Importances from Tuned Random Forest (Prefixed Names)')
plt.tight_layout()
plot_filename = os.path.join(MODELS_DIR, f"{datetime.now().strftime('%Y%m%d_%H%M%S')}_feature_importances_prefixed.png")
try:
    plt.savefig(plot_filename)
    logging.info(f"Saved feature importance plot to {plot_filename}")
except Exception as e:
    logging.error(f"Failed to save feature importance plot: {e}", exc_info=True)
plt.close()

# --- 5. Apply Feature Selection using SelectFromModel ---
logging.info("Applying feature selection using SelectFromModel with tuned Random Forest.")
# Using median importance as threshold. Can also use a specific float value or "mean", "1.25*mean", etc.
selector = SelectFromModel(tuned_rf_model, threshold='median', prefit=True) # prefit=True as model is already trained

X_train_selected = selector.transform(X_train_processed_df) # X_train_processed_df now has prefixed names
selected_feature_names = list(X_train_processed_df.columns[selector.get_support(indices=True)]) # Get names directly from df columns

logging.info(f"Original number of features: {X_train_processed_df.shape[1]}")
logging.info(f"Number of features selected: {X_train_selected.shape[1]}")
logging.info(f"Selected feature names (first 20): {selected_feature_names[:20]}")

# Save the selected feature names (these will be prefixed)
selected_features_filename = f"{datetime.now().strftime('%Y%m%d_%H%M%S')}_selected_feature_names_prefixed.txt"
selected_features_path = os.path.join(MODELS_DIR, selected_features_filename)
with open(selected_features_path, 'w') as f:
    for name in selected_feature_names:
        f.write(f"{name}\n")
logging.info(f"Saved selected (prefixed) feature names to {selected_features_path}")

# Create DataFrame for selected features (maintaining prefixed names)
X_train_selected_df = pd.DataFrame(X_train_selected, columns=selected_feature_names, index=X_train_processed_df.index)
logging.info(f"X_train_selected_df created with shape: {X_train_selected_df.shape}")

# --- 6. Retrain and Evaluate Models on Reduced Feature Set ---
logging.info("Retraining and evaluating models on the reduced (prefixed) feature set.")

models_to_re_evaluate = {}
# Load tuned Gradient Boosting model (from Step 3.1)
tuned_gb_model_path = find_latest_file_by_keyword(MODELS_DIR, "Gradient_Boosting_tuned", endswith=".joblib") # Ensure this function is available
if tuned_gb_model_path:
    logging.info(f"Loading tuned Gradient Boosting model from: {tuned_gb_model_path}")
    try:
        tuned_gb_model = joblib.load(tuned_gb_model_path)
        # We will use the already loaded tuned_rf_model for Random Forest
        models_to_re_evaluate["Tuned Random Forest (Selected Prefixed Feats)"] = tuned_rf_model
        models_to_re_evaluate["Tuned Gradient Boosting (Selected Prefixed Feats)"] = tuned_gb_model
    except Exception as e:
        logging.error(f"Error loading tuned Gradient Boosting model: {e}", exc_info=True)
        # Fallback to only RF if GB fails to load
        if "Tuned Random Forest (Selected Prefixed Feats)" not in models_to_re_evaluate:
             models_to_re_evaluate["Tuned Random Forest (Selected Prefixed Feats)"] = tuned_rf_model
else:
    logging.warning("Tuned Gradient Boosting model not found. Only re-evaluating Random Forest.")
    models_to_re_evaluate["Tuned Random Forest (Selected Prefixed Feats)"] = tuned_rf_model


cv_results_selected_features = []
kf = KFold(n_splits=5, shuffle=True, random_state=42)

for model_name, model_instance in models_to_re_evaluate.items():
    logging.info(f"Cross-validating {model_name} on selected (prefixed) features...")
    start_time = time.time()
    # The model_instance is already tuned. We are evaluating it on the new feature set.
    # If the model was tuned on more features, its internal state (e.g. feature importances for RF)
    # might not directly map if it relies on indices. However, scikit-learn models' fit/predict
    # methods generally work with data that has the features they were trained on or a subset if applicable (though SelectFromModel handles this).
    # For cross-validation here, we are effectively retraining on the selected subset.
    # To be absolutely rigorous, one might re-tune hyperparameters on the selected features,
    # but for now, we use the previously tuned parameters.

    # To ensure the model uses only the selected features for this CV, we pass X_train_selected_df
    cv_scores = cross_val_score(model_instance, X_train_selected_df, y_train_series, cv=kf, scoring='r2', n_jobs=-1)
    cv_time = time.time() - start_time
    
    logging.info(f"{model_name} - CV R2 scores: {cv_scores}")
    logging.info(f"{model_name} - Mean CV R2: {np.mean(cv_scores):.4f} (+/- {np.std(cv_scores):.4f})")
    logging.info(f"{model_name} - CV Time: {cv_time:.2f}s")
    
    cv_results_selected_features.append({
        "Model": model_name,
        "Mean CV R2": np.mean(cv_scores),
        "Std CV R2": np.std(cv_scores),
        "CV Time (s)": cv_time,
        "Num Features": X_train_selected_df.shape[1]
    })

# --- 7. Results Summary for Selected Features ---
if cv_results_selected_features:
    results_selected_df = pd.DataFrame(cv_results_selected_features)
    results_selected_df = results_selected_df.sort_values(by="Mean CV R2", ascending=False).reset_index(drop=True)
    print("\n--- Model Performance on Selected (Prefixed) Features (Cross-Validation) ---")
    print(results_selected_df)
    
    # You might want to compare these scores to the scores from Step 3.1 (on all features)
    # logging.info("Compare with scores from Step 3.1 (all features if available).")
else:
    print("\nNo model evaluation results on selected features to display.")

logging.info("Step 3.2: Feature Selection script (Updated for prefixed names) finished.")
````

````python
# Step 3.3: Advanced Ensemble Techniques (Stacking)
#
# This script focuses on:
# 1. Loading data and applying preprocessing, then selecting features.
# 2. Loading best tuned base models (Random Forest, Gradient Boosting).
# 3. Training a StackingRegressor using these base models.
# 4. Evaluating the stacked model on the test set.

import pandas as pd
import numpy as np
import os
import logging
import time
from datetime import datetime
import joblib

# Scikit-learn imports
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Preprocessor components
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - [%(filename)s:%(lineno)d] - %(message)s', force=True)

logging.info("Libraries imported and logging configured.")
logging.info("Starting Step 3.3: Stacking Ensemble.")

# --- 1. Configuration and Data Loading ---
logging.info("Setting up configuration for data loading.")
try:
    SCRIPT_DIR = os.path.dirname(os.path.abspath(__file__))
    WORKSPACE_ROOT = os.path.abspath(os.path.join(SCRIPT_DIR, '../../../'))
    if os.path.basename(WORKSPACE_ROOT) != 'Alpha GO': # Replace if your workspace dir is different
        logging.warning(f"Auto-detected WORKSPACE_ROOT might be incorrect: {WORKSPACE_ROOT}. Falling back.")
        raise ValueError("Incorrect workspace root detected")
except Exception:
    logging.warning(f"Could not auto-detect WORKSPACE_ROOT correctly. Using hardcoded path as fallback.")
    WORKSPACE_ROOT = "D:/SBT-JAPAN/Alpha GO" #! IMPORTANT: UPDATE THIS TO YOUR ABSOLUTE WORKSPACE PATH

DATA_DIR = os.path.join(WORKSPACE_ROOT, "dataset-consolidation", "data")
FEATURE_ENGINEERED_DATA_DIR = os.path.join(DATA_DIR, "processed")
MODELS_DIR = os.path.join(WORKSPACE_ROOT, "dataset-consolidation", "models") # base_models and ensemble
BASE_MODELS_DIR = os.path.join(MODELS_DIR, "base_models")
ENSEMBLE_MODELS_DIR = os.path.join(MODELS_DIR, "ensemble")

os.makedirs(ENSEMBLE_MODELS_DIR, exist_ok=True)

logging.info(f"WORKSPACE_ROOT set to: {WORKSPACE_ROOT}")
logging.info(f"FEATURE_ENGINEERED_DATA_DIR set to: {FEATURE_ENGINEERED_DATA_DIR}")
logging.info(f"BASE_MODELS_DIR set to: {BASE_MODELS_DIR}")
logging.info(f"ENSEMBLE_MODELS_DIR set to: {ENSEMBLE_MODELS_DIR}")


def find_latest_file_by_keyword(directory, keyword, endswith=".csv"):
    # (Assuming this function is defined as in previous scripts)
    logging.debug(f"Searching for files with keyword '{keyword}' ending with '{endswith}' in directory: {directory}")
    try:
        if not os.path.isdir(directory):
            logging.error(f"Directory not found: {directory}")
            return None
        files = [f for f in os.listdir(directory) if keyword in f and f.endswith(endswith) and os.path.isfile(os.path.join(directory, f))]
        if not files:
            logging.error(f"No files found with keyword '{keyword}' ending with '{endswith}' in {directory}")
            return None
        files.sort(reverse=True)
        latest_file = files[0]
        logging.debug(f"Found latest file for keyword '{keyword}': {latest_file}")
        return os.path.join(directory, latest_file)
    except Exception as e:
        logging.error(f"Error finding latest file for '{keyword}': {e}", exc_info=True)
        return None

# Load original full feature-engineered data
latest_fe_file_path = find_latest_file_by_keyword(FEATURE_ENGINEERED_DATA_DIR, "_cars_feature_engineered.csv")
if not latest_fe_file_path:
    logging.critical(f"No feature engineered CSV file found in {FEATURE_ENGINEERED_DATA_DIR}. Exiting.")
    exit()
logging.info(f"Loading original feature-engineered data from: {latest_fe_file_path}")
df_original_fe = pd.read_csv(latest_fe_file_path)

TARGET_COLUMN_ORIGINAL = 'price_log'
if TARGET_COLUMN_ORIGINAL not in df_original_fe.columns:
    logging.error(f"Target column '{TARGET_COLUMN_ORIGINAL}' not found in {latest_fe_file_path}. Exiting.")
    exit()

X_original_fe = df_original_fe.drop(columns=[TARGET_COLUMN_ORIGINAL])
y_original_fe = df_original_fe[TARGET_COLUMN_ORIGINAL]

# Perform 80/20 train-test split
X_train_unprocessed, X_test_unprocessed, y_train_series, y_test_series = train_test_split(
    X_original_fe, y_original_fe, test_size=0.2, random_state=42
)
logging.info(f"X_train_unprocessed shape: {X_train_unprocessed.shape}, X_test_unprocessed shape: {X_test_unprocessed.shape}")
logging.info(f"y_train_series shape: {y_train_series.shape}, y_test_series shape: {y_test_series.shape}")


# --- 2. Load Selected Feature Names ---
selected_features_path = find_latest_file_by_keyword(BASE_MODELS_DIR, "_selected_feature_names.txt", endswith=".txt")
if not selected_features_path:
    logging.critical("'_selected_feature_names.txt' not found. Please run feature selection script first. Exiting.")
    exit()
logging.info(f"Loading selected feature names from: {selected_features_path}")
with open(selected_features_path, 'r') as f:
    selected_feature_names = [line.strip() for line in f if line.strip()]
logging.info(f"Loaded {len(selected_feature_names)} selected feature names.")
if not selected_feature_names:
    logging.error("Selected feature names list is empty. Exiting.")
    exit()


# --- 3. Preprocess Data and Apply Feature Selection ---
# Re-define preprocessor (must match the one used for getting transformed_feature_names in previous script)
numeric_features = X_train_unprocessed.select_dtypes(include=np.number).columns.tolist()
categorical_features = X_train_unprocessed.select_dtypes(include='object').columns.tolist()
preprocessor = None
if 'annual_insurance' in numeric_features:
    numeric_transformer_main = Pipeline(steps=[('scaler', StandardScaler())])
    numeric_transformer_insurance = Pipeline(steps=[
        ('imputer', IterativeImputer(random_state=42)),
        ('scaler', StandardScaler())
    ])
    other_numeric_features = [col for col in numeric_features if col != 'annual_insurance']
    categorical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='constant', fill_value='missing_value_runtime')), # Should not be hit if FE worked
        ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
    ])
    preprocessor = ColumnTransformer(
        transformers=[
            ('num_insurance', numeric_transformer_insurance, ['annual_insurance']),
            ('num_main', numeric_transformer_main, other_numeric_features),
            ('cat', categorical_transformer, categorical_features)
        ], remainder='passthrough'
    )
else: # Simplified case if annual_insurance isn't handled specially (shouldn't happen based on our data)
    logging.error("'annual_insurance' special handling in preprocessor definition is expected. Check logic. Exiting.")
    exit()

logging.info("Fitting preprocessor on X_train_unprocessed...")
preprocessor.fit(X_train_unprocessed)

# Get all transformed feature names
try:
    ohe_feature_names = preprocessor.named_transformers_['cat'].named_steps['onehot'].get_feature_names_out(categorical_features)
    all_transformed_feature_names = []
    if 'annual_insurance' in numeric_features: # Assuming this was the structure
         all_transformed_feature_names.extend(['annual_insurance'])
         all_transformed_feature_names.extend(other_numeric_features)
    else:
        all_transformed_feature_names.extend(numeric_features)
    all_transformed_feature_names.extend(ohe_feature_names)
    logging.info(f"Generated {len(all_transformed_feature_names)} total transformed feature names for consistency check.")
except Exception as e:
    logging.error(f"Could not get all transformed feature names from preprocessor: {e}. Exiting.")
    exit()

logging.info("Transforming X_train_unprocessed...")
X_train_processed_all_feats_np = preprocessor.transform(X_train_unprocessed)
X_train_processed_all_feats_df = pd.DataFrame(X_train_processed_all_feats_np, columns=all_transformed_feature_names, index=X_train_unprocessed.index)

logging.info("Transforming X_test_unprocessed...")
X_test_processed_all_feats_np = preprocessor.transform(X_test_unprocessed)
X_test_processed_all_feats_df = pd.DataFrame(X_test_processed_all_feats_np, columns=all_transformed_feature_names, index=X_test_unprocessed.index)

# Filter to keep only selected features
logging.info(f"Filtering datasets to keep only the {len(selected_feature_names)} selected features.")
# Ensure all selected_feature_names are present in the transformed columns
missing_selected_feats = [name for name in selected_feature_names if name not in X_train_processed_all_feats_df.columns]
if missing_selected_feats:
    logging.error(f"Some selected features are not in the transformed columns: {missing_selected_feats}. Exiting.")
    exit()

X_train_selected_df = X_train_processed_all_feats_df[selected_feature_names]
X_test_selected_df = X_test_processed_all_feats_df[selected_feature_names]

logging.info(f"X_train_selected_df shape: {X_train_selected_df.shape}")
logging.info(f"X_test_selected_df shape: {X_test_selected_df.shape}")


# --- 4. Load Tuned Base Models ---
logging.info("Loading tuned base models...")
tuned_rf_model_path = find_latest_file_by_keyword(BASE_MODELS_DIR, "Random_Forest_tuned", endswith=".joblib")
tuned_gb_model_path = find_latest_file_by_keyword(BASE_MODELS_DIR, "Gradient_Boosting_tuned", endswith=".joblib")

base_estimators = []
if tuned_rf_model_path:
    logging.info(f"Loading Tuned Random Forest from {tuned_rf_model_path}")
    tuned_rf_model = joblib.load(tuned_rf_model_path)
    base_estimators.append(('random_forest', tuned_rf_model))
else:
    logging.warning("Tuned Random Forest model not found. It will be excluded from the stack.")

if tuned_gb_model_path:
    logging.info(f"Loading Tuned Gradient Boosting from {tuned_gb_model_path}")
    tuned_gb_model = joblib.load(tuned_gb_model_path)
    base_estimators.append(('gradient_boosting', tuned_gb_model))
else:
    logging.warning("Tuned Gradient Boosting model not found. It will be excluded from the stack.")

if not base_estimators:
    logging.critical("No base models loaded. Cannot proceed with stacking. Exiting.")
    exit()

logging.info(f"Base estimators for stacking: {[name for name, _ in base_estimators]}")


# --- 5. Define and Train Stacking Regressor ---
logging.info("Defining and training StackingRegressor...")
# Using LinearRegression as the meta-regressor.
# CV for the meta-regressor is to generate out-of-fold predictions from base models for training the final_estimator.
# The final_estimator itself is then trained on these predictions.
stacking_regressor = StackingRegressor(
    estimators=base_estimators,
    final_estimator=LinearRegression(),
    cv=KFold(n_splits=5, shuffle=True, random_state=42), # CV for base model predictions
    n_jobs=-1,
    passthrough=False # Only use predictions of base models as input to final_estimator
)

start_time = time.time()
logging.info("Fitting the StackingRegressor on X_train_selected_df...")
stacking_regressor.fit(X_train_selected_df, y_train_series)
fit_time = time.time() - start_time
logging.info(f"StackingRegressor training completed in {fit_time:.2f} seconds.")


# --- 6. Evaluate Stacked Model ---
logging.info("Evaluating StackingRegressor on the test set (X_test_selected_df)...")
try:
    test_predictions = stacking_regressor.predict(X_test_selected_df)
    
    r2 = r2_score(y_test_series, test_predictions)
    mae = mean_absolute_error(y_test_series, test_predictions)
    rmse = np.sqrt(mean_squared_error(y_test_series, test_predictions))

    logging.info(f"--- Stacking Regressor Performance on Test Set ---")
    logging.info(f"R-squared: {r2:.4f}")
    logging.info(f"MAE: {mae:.4f}")
    logging.info(f"RMSE: {rmse:.4f}")
    print(f"\n--- Stacking Regressor Performance on Test Set ---")
    print(f"R-squared: {r2:.4f}")
    print(f"MAE: {mae:.4f}")
    print(f"RMSE: {rmse:.4f}")

except Exception as e:
    logging.error(f"Error during StackingRegressor evaluation: {e}", exc_info=True)
    r2, mae, rmse = np.nan, np.nan, np.nan


# --- 7. Save Stacked Model ---
if not np.isnan(r2): # Only save if evaluation was successful
    try:
        stacked_model_filename = f"{datetime.now().strftime('%Y%m%d_%H%M%S')}_StackingRegressor_RF_GB_Linear.joblib"
        stacked_model_path = os.path.join(ENSEMBLE_MODELS_DIR, stacked_model_filename)
        joblib.dump(stacking_regressor, stacked_model_path)
        logging.info(f"Saved StackingRegressor model to {stacked_model_path}")
    except Exception as e_save:
        logging.error(f"Error saving StackingRegressor model: {e_save}")
else:
    logging.warning("StackingRegressor evaluation failed or produced NaN metrics. Model not saved.")


logging.info("Step 3.3: Stacking Ensemble script finished.")
````
